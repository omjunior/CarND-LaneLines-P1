## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

# untouched
def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# untouched
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

# untouched
def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

# added
def bilateral_filter(img, d, sigmaColor, sigmaSpace):
    """Applies a Bilateral Filter"""
    return cv2.bilateralFilter(img, d, sigmaColor, sigmaSpace)

# untouched
def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

# untouched
def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)


# Global variable to store previous frames' info
prevFrames = None            
            
# added
def draw_lines_extrapolating(img, lines, color=[255, 0, 0], thickness=2, video=False):
    
    global prevFrames
    
    # Compute parameters of each line found (m, b, len)
    parameters = np.zeros((lines.shape[0], 3), dtype=np.float64)
    for i, line in enumerate(lines):
        for x1, y1, x2, y2 in line:
            m = (y2-y1)/(x2-x1)
            b = y1 - m*x1
            l = math.sqrt((y2-y1)**2 + (x2-x1)**2)
            parameters[i, 0] = m
            parameters[i, 1] = b
            parameters[i, 2] = l
        
    # Divide the lines by slope signal and merge them by averaging m and b
    # using their length as weight
    # (discard inclinations under 30 deg)
    
    p_slope_mask = np.logical_and(parameters[:,0] >= 0.5, parameters[:,0] < float('inf'))
    n_slope_mask = np.logical_and(parameters[:,0] <= -0.5, parameters[:,0] > float('-inf'))
    main_lines = np.zeros((2, 1, 4), dtype=np.int64)

    m1, b1, m2, b2 = float('NaN'), float('NaN'), float('NaN'), float('NaN')
    if (sum(p_slope_mask) > 0):
        m1, b1 = (sum(parameters[p_slope_mask, 0:2] * parameters[p_slope_mask, None, 2])
                / sum(parameters[p_slope_mask, 2]))
    if (sum(n_slope_mask) > 0):
        m2, b2 = (sum(parameters[n_slope_mask, 0:2] * parameters[n_slope_mask, None, 2])
                / sum(parameters[n_slope_mask, 2]))

    currentFrame = np.array([m1, b1, m2, b2])
    
    if video:
        # defines the number of frames to look (including current)
        lookBack = 12 # 25 fps
        smoothness = 2
        weights = np.array([1/x for x in range(smoothness, lookBack+smoothness)])
        weights = weights / sum(weights)
        if (prevFrames is None):
            # initialize with first frame
            prevFrames = np.tile(currentFrame, lookBack)
            prevFrames = prevFrames.reshape(lookBack, 4)
        # if NaN, keep last frame
        currentFrame = [p if math.isnan(c) else c for c, p in zip(currentFrame, prevFrames[0])]
        # update with current
        prevFrames = np.roll(prevFrames, 1, axis=0)
        prevFrames[0, :] = currentFrame
        # compute weighted avg
        currentFrame = sum(prevFrames * weights[:, None])
        
    main_lines[0, 0, :] = extrapolate_line(currentFrame[0], currentFrame[1],
                                           img.shape[0]*.60, img.shape[0])
    main_lines[1, 0, :] = extrapolate_line(currentFrame[2], currentFrame[3],
                                           img.shape[0]*.60, img.shape[0])
    
    # Debug: Uncomment both lines below to get raw lines
    #debug_slope = np.logical_and(abs(parameters[:,0]) > 0.5, abs(parameters[:,0]) < float('inf'))
    #main_lines = lines[debug_slope]

    draw_lines(img, main_lines, color, thickness)

# added
def extrapolate_line(m, b, y_min, y_max):
    """ Extrapolate a line given m and b, for y values between y_min and y_max """
    if (math.isnan(m) or math.isnan(b)):
        return [0, 0, 0, 0]
    return [int(round((y_min-b)/m)), int(round(y_min)), int(round((y_max-b)/m)), int(round(y_max))]
    
# changed            
def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap, video=False):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines_extrapolating(line_img, lines, thickness=5, video=video)
    return line_img
   
# Python 3 has support for cool math symbols.
# untouched
def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

````python
def pipeline(img, video=False):
    """ Find the left and right lines on the road """
    
    # Skip grayscaling here!
    #image = grayscale(img)
    image = img
    
    # Apply a bilateral filter
    # Able to aply a more agressive smoothing keeping edges
    image = bilateral_filter(image, 9, 75, 75)
    # Apply gaussian blur to smooth thin sharp edges
    image = gaussian_blur(image, 5)
    
    # Apply a Canny detection for each color channel and max them together
    #image = canny(image, 100, 200)
    image_r = canny(image[:,:,0], 100, 200)
    image_g = canny(image[:,:,1], 100, 200)
    image_b = canny(image[:,:,2], 100, 200)
    image = np.fmax(image_r, image_g, image_b)
    
    # Select ROI based on the image size (constant proportion)
    image = region_of_interest(image, np.array([[(image.shape[1]*.45, image.shape[0]*.60),
                                                 (image.shape[1]*.55, image.shape[0]*.60),
                                                 (image.shape[1]*.95, image.shape[0]),
                                                 (image.shape[1]*.05, image.shape[0])]], dtype=np.int32))

    # Compute Hough Lines over the ROI of the image
    lines_img = hough_lines(image, 1, np.pi/180, 10, 30, 15, video=video)

    # Overlay lines
    return weighted_img(lines_img, img)


# Loop through files in test_images dir
plots = plt.subplots((len(test_files)+1)//2, 2, figsize=(16, 12))[1]
for i, file in enumerate(test_files):
    image = mpimg.imread('test_images/' + file)
    image = pipeline(image)
    plots[i//2, i%2].imshow(image)
    mpimg.imsave('test_images_output/' + file, image, format='jpg')
````
